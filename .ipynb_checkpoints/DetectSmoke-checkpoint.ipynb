```python
# Contour Detector Function
def detect_smoke_from_image(image_path, threshold_type='mean'):
    # Load the image
    image = cv2.imread(image_path)
    
    if image is None:
        print(f"Error loading image: {image_path}")
        return None

    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Determine threshold value based on the selected type
    if threshold_type == 'mean':
        threshold_value = np.mean(gray_image)
    elif threshold_type == 'mode':
        threshold_value = np.median(gray_image)
    elif threshold_type == 'median':
        threshold_value = np.median(gray_image)

    # Create a binary mask based on the threshold
    binary_mask = gray_image < threshold_value

    # Apply morphological operations to remove noise
    kernel = np.ones((5, 5), np.uint8)
    binary_mask = cv2.morphologyEx(binary_mask.astype(np.uint8), cv2.MORPH_OPEN, kernel)

    # Find contours in the binary mask
    contours, _ = cv2.findContours(binary_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Draw contours on the original image
    output_image = image.copy()
    cv2.drawContours(output_image, contours, -1, (0, 255, 0), 2)

    return output_image
# Display the images with smoke detection
folder_path = os.path.join(os.getcwd(), 'Images')
image_files = os.listdir(folder_path)

plt.figure(figsize=(20, 12))
for i, image_file in enumerate(image_files, 1):
    image_path = os.path.join(folder_path, image_file)
    output_image = detect_smoke_from_image(image_path)

    if output_image is not None:
        plt.subplot(2, 5, i)
        plt.imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
        plt.title(f'Image {i}')
        plt.axis('off')
plt.show()
```

Smoke Detection From Video 

### Video Analyser Function

```python
# Function to detect smoke from video frames
def detect_smoke_from_video(frame_id, frame, time, threshold_type='mean'):
    # Convert the frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Determine threshold value based on the selected type
    if threshold_type == 'mean':
        threshold_value = np.mean(gray_frame)
    elif threshold_type == 'mode':
        threshold_value = np.median(gray_frame)
    elif threshold_type == 'median':
        threshold_value = np.median(gray_frame)

    # Create a binary mask based on the threshold
    binary_mask = gray_frame < threshold_value

    # Apply morphological operations to remove noise
    kernel = np.ones((5,5), np.uint8)
    binary_mask = cv2.morphologyEx(binary_mask.astype(np.uint8), cv2.MORPH_OPEN, kernel)

    # Find contours in the binary mask
    contours, _ = cv2.findContours(binary_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Draw contours on the original frame
    output_frame = frame.copy()
    cv2.drawContours(output_frame, contours, -1, (0, 255, 0), 2)

    # Calculate statistics on detected smoke regions
    num_smoke_regions = len(contours)
    smoke_areas = [cv2.contourArea(contour) for contour in contours]
    avg_smoke_area = np.mean(smoke_areas) if smoke_areas else 0

    # Calculate statistical measures
    mean_value = np.mean(gray_frame)
    median_value = np.median(gray_frame)
    mode_value = np.argmax(np.bincount(gray_frame.flatten()))
    std_dev_value = np.std(gray_frame)

    # Append the statistical measures to the lists
    means.append(mean_value)
    medians.append(median_value)
    modes.append(mode_value)
    std_devs.append(std_dev_value)

    # Append the statistics related to smoke regions
    num_smoke_regions_list.append(num_smoke_regions)
    avg_smoke_area_list.append(avg_smoke_area)

    # Determine if smoke is detected and append label
    label = 1 if num_smoke_regions > 0 else 0
    labels.append(label)

    # Extract RGB values if smoke regions are detected
    if num_smoke_regions > 0:
        red_channel = frame[:,:,0]
        green_channel = frame[:,:,1]
        blue_channel = frame[:,:,2]

        # Calculate average RGB values
        avg_red = np.mean(red_channel)
        avg_green = np.mean(green_channel)
        avg_blue = np.mean(blue_channel)
        # Append RGB values to lists
        reds.append(avg_red)
        greens.append(avg_green)
        blues.append(avg_blue)
    else:
        # If no smoke regions are detected, append NaN values
        reds.append(np.nan)
        greens.append(np.nan)
        blues.append(np.nan)
    # Annotate the frame with statistics
    cv2.putText(output_frame, f"Frame ID: {frame_id}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
    cv2.putText(output_frame, f"Time: {time:.2f}s", (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
    cv2.putText(output_frame, f"Mean: {mean_value:.2f}", (10, 90), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
    cv2.putText(output_frame, f"Median: {median_value:.2f}", (10, 120), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
    cv2.putText(output_frame, f"Mode: {mode_value}", (10, 150), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
    cv2.putText(output_frame, f"Std Dev: {std_dev_value:.2f}", (10, 180), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
    cv2.putText(output_frame, f"# Smoke Regions: {num_smoke_regions}", (10, 210), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
    cv2.putText(output_frame, f"Avg Smoke Area: {avg_smoke_area:.2f}", (10, 240),
    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)

    return output_frame
```

Convert Videos to Frames

```python
def process_video_frames(video_path):
    cap = cv2.VideoCapture(video_path)
    frame_rate = cap.get(cv2.CAP_PROP_FPS)
    # Initialize frame id
    frame_id = 1
    # VideoCapture object creation
    cap = cv2.VideoCapture(video_path)
    # Process the video frame by frame
    while True:
        # Read a frame from the video
        ret, frame = cap.read()

        if not ret:
            break

        # Calculate the current time in seconds
        time = frame_id / frame_rate

        # Process the frame
        output_frame = detect_smoke_from_video(frame_id, frame, time)

        # Display the frame
        cv2.imshow('Frame', output_frame)

        # Add a delay corresponding to the video frame rate
        if cv2.waitKey(int(30 / frame_rate)) & 0xFF == ord('q'):
            break

        frame_id += 1

    # Release the VideoCapture and close all windows
    cap.release()
    cv2.destroyAllWindows() 
```

### Conversion to Dataframe

```python

def create_dataframe_from_analysis_output( means, medians, modes, std_devs, num_smoke_regions_list, avg_smoke_area_list, reds, greens, blues, labels):
    # Initialize frame id
    frame_id = 1
    # DataFrame columns
    columns = ['Frame ID', 'Mean', 'Median', 'Mode', 'Std Dev', 'Num Smoke Regions', 'Avg Smoke Area', 'Avg Red', 'Avg Green', 'Avg Blue', 'Label']

    # Create a DataFrame from the statistical measures and RGB values
    data = {
#         'Frame ID': range(1, frame_id),
        'Mean': means,
        'Median': medians,
        'Mode': modes,
        'Std Dev': std_devs,
        'Num Smoke Regions': num_smoke_regions_list,
        'Avg Smoke Area': avg_smoke_area_list,
        'Avg Red': reds,
        'Avg Green': greens,
        'Avg Blue': blues,
        'Label': labels
    }
    df = pd.DataFrame(data)
    return df
```

## Machine Learning Approach to Smoke Detection

```python
# ML PKGS
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix,precision_score,recall_score
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier


# Deep Learning KGS
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
import tensorflow.keras.backend as K
```

#### Data Splitting into train label and test labels

```python
# Prepare data for meal prediction
X = dataset.drop('Is_Smoke',axis = 1)
y = dataset['Is_Smoke']
```

### Feature Engineering

```python
# Scale features
scaler = StandardScaler()
X = scaler.fit_transform(X)
X.shape,y.shape
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.shape,X_test.shape
```

### Configure Model To Benchmark

```python
models = {
    'SVM': SVC(kernel='linear'),
    'Random Forest': RandomForestClassifier(),
    'Logistic Regression': LogisticRegression(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'K-Nearest Neighbors': KNeighborsClassifier()
}
model_accuracies = {}
```

```python
# Looping the models from the pipeline
for model_name, model in models.items():
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred) * 100
    
    model_accuracies[model_name] = accuracy
    print(f"{model_name} Accuracy: {accuracy:.4f}%")
```

### Model Sensitivity Analysis

```python
# Create dictionaries to store precision, recall, and accuracy of each model
model_precision = {}
model_recall = {}
model_accuracy = {}
results = {}
# Create pipelines for each model and evaluate
for model_name, model in models.items():
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])
    
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    
    model_precision[model_name] = precision
    model_recall[model_name] = recall
    model_accuracy[model_name] = accuracy
    results[model_name] = {'accuracy': accuracy, 'confusion_matrix': conf_matrix}
    print(f"{model_name} - Precision: {precision:.4f}, Recall: {recall:.4f}, Accuracy: {accuracy:.4f}")
    
```

```python
plt.figure(figsize=(8, 8))
barplot = sns.barplot(x=list(results.keys()), y=[result['accuracy'] for result in results.values()])
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Comparison of Model Accuracies')

# Add annotations to the bar plot
for idx, result in enumerate(results.values()):
    barplot.text(idx, result['accuracy'] + 0.01, f'{result["accuracy"]:.4f}', ha='center', color='black')
plt.show()
```
